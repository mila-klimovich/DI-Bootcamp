```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson

# Create figure and subplots
plt.figure(figsize=(15, 5))

# -------------------------------
# 1. Gaussian (Normal) Distribution
# -------------------------------
plt.subplot(1, 3, 1)
x = np.linspace(-0.3, 0.3, 500)
mu = 0
sigma = 0.05
pdf = norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, pdf, color='teal')
plt.title('Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)

# -------------------------------
# 2. Binomial Distribution
# -------------------------------
plt.subplot(1, 3, 2)
n = 10    # number of trials
p = 0.5   # probability of success
x = np.arange(0, n+1)
pmf_binom = binom.pmf(x, n, p)
plt.bar(x, pmf_binom, color='teal')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass')
plt.grid(True)

# -------------------------------
# 3. Poisson Distribution
# -------------------------------
plt.subplot(1, 3, 3)
lambda_ = 3  # average number of events
x = np.arange(0, 9)
pmf_poisson = poisson.pmf(x, mu=lambda_)
plt.bar(x, pmf_poisson, color='teal')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability Mass')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np

# Simulate two groups with the same mean (for now)
group1 = stats.norm.rvs(loc=5, scale=10, size=100)
group2 = stats.norm.rvs(loc=5, scale=10, size=100)

# Perform independent two-sample T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Output
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
from scipy import stats

# Sample data from three different groups
group1 = [20, 21, 19, 22, 24]
group2 = [28, 30, 27, 26, 29]
group4 = [5, 5, 3, 1, 2]
group3 = [33, 35, 34, 31, 32]

# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(group1, group2, group3, group4)
print("F-value:", f_value)
print("P-value:", p_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Define the dataset
x = np.array([2, 4, 6, 8, 10])
y = np.array([3, 5, 7, 9, 11])

# Step 2: Apply linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print the regression results
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)

# Step 3: Plot the data points and regression line
plt.figure(figsize=(7, 4))
plt.scatter(x, y, color='blue', label='Data points')  # Original data
plt.plot(x, slope * x + intercept, color='red', label='Regression line')  # Fitted line
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Example data
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)

# Calculate Pearson correlation
r, _ = pearsonr(x, y)

# Plot
plt.scatter(x, y)
plt.title(f"Pearson Correlation (r = {r:.2f})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

r
```

```python
from scipy import stats

x = np.array([2, 4, 6, 8, 10])
y = np.array([4, 8, 12, 16, 20])

# Regression
slope, intercept, r, p, stderr = stats.linregress(x, y)

# Plot
plt.scatter(x, y, label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Regression line')
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate covariance matrix
cov_matrix = np.cov(x, y)
cov_xy = cov_matrix[0, 1]

# Plot
plt.scatter(x, y)
plt.title(f"Covariance (Cov = {cov_xy:.2f})")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
cov_xy
```

```python
import pandas as pd

data = pd.DataFrame({
    'Age': [25, 40, 30, 22, 35],
    'Monthly_Income': [3200, 5800, 4500, 2800, 5000],
    'Online_Purchase_Frequency': [15, 8, 12, 18, 10],
    'Hours_Social_Media': [30, 15, 25, 40, 20]
})
```

```python
from scipy.stats import pearsonr

corr, _ = pearsonr(data['Monthly_Income'], data['Age'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')
```

```python
from scipy.stats import spearmanr

corr, _ = spearmanr(data['Age'], data['Hours_Social_Media'])
print(f'Spearman Rank Correlation: {corr:.2f}')
```

```python
import numpy as np

covariance = np.cov(data['Age'], data['Monthly_Income'])[0, 1]
print(f'Covariance: {covariance:.2f}')
```

```python
import matplotlib.pyplot as plt

# Monthly Income vs. Online Purchase Frequency
plt.scatter(data['Monthly_Income'], data['Online_Purchase_Frequency'])
plt.title('Monthly Income vs. Online Purchase Frequency')
plt.xlabel('Monthly Income')
plt.ylabel('Online Purchase Frequency')
plt.grid(True)
plt.show()

# Age vs. Hours Spent on Social Media
plt.scatter(data['Age'], data['Hours_Social_Media'])
plt.title('Age vs. Hours Spent on Social Media')
plt.xlabel('Age')
plt.ylabel('Hours on Social Media')
plt.grid(True)
plt.show()
```
