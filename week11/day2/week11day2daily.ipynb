```python
# Data Cleaning and Preparation

# Check for missing values in all columns
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Handling Missing Values Based on Dataset Inspection

# 1. 'other_fuel1', 'other_fuel2', 'other_fuel3' - many missing values, optional columns, we leave them as is or can drop
# 2. 'commissioning_year' - important for time analysis, fill missing with 0 or leave NaN depending on the task
# 3. Generation columns - can fill with 0, assuming missing means no data reported
# 4. 'owner', 'source', 'url', 'geolocation_source', 'wepp_id', 'year_of_capacity_data' - optional, we keep NaN for now
# 5. Estimated generation columns - will fill with 0

# Fill missing generation columns with 0
generation_cols = [col for col in df.columns if 'generation_gwh' in col and 'estimated' not in col]
df[generation_cols] = df[generation_cols].fillna(0)

# Fill estimated generation columns with 0
estimated_cols = [col for col in df.columns if 'estimated_generation_gwh' in col]
df[estimated_cols] = df[estimated_cols].fillna(0)

# Fill 'commissioning_year' missing with 0 (alternative: drop rows or keep NaN depending on analysis)
df['commissioning_year'] = df['commissioning_year'].fillna(0)

# Re-check for missing values
print("Missing values after basic cleaning:\n", df.isnull().sum())
```

```python
# Ensure Numerical Types with NumPy

import numpy as np

# Columns to check/convert to numeric (should be float, but we confirm)
numeric_cols = ['capacity_mw', 'latitude', 'longitude', 'commissioning_year'] + generation_cols + estimated_cols

# Convert using NumPy where necessary
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Confirm types
print("\nData types after conversion:\n")
print(df[numeric_cols].dtypes)
```

```python
# Exercise: Exploratory Data Analysis - Summary Statistics for Numerical Columns

# Summary statistics using pandas describe() gives count, mean, std, min, max, quartiles
print(df.describe())

# Additional specific statistics for capacity_mw
mean_capacity = df['capacity_mw'].mean()
median_capacity = df['capacity_mw'].median()
std_capacity = df['capacity_mw'].std()

print(f"\nCapacity (MW) - Mean: {mean_capacity:.2f}, Median: {median_capacity:.2f}, Std Dev: {std_capacity:.2f}")
```

```python
# Exercise: Distribution of Power Plants by Country

# Count power plants by country_long (full country name)
country_counts = df['country_long'].value_counts()

# Display top 10 countries with most power plants
print("Top 10 countries by number of power plants:\n", country_counts.head(10))
```

```python
# Exercise: Distribution of Power Plants by Fuel Type

# Count power plants by primary fuel type
fuel_counts = df['primary_fuel'].value_counts()

# Display all fuel types
print("Distribution of power plants by fuel type:\n", fuel_counts)
```

```python
# Exercise: Statistical Analysis of Power Output by Fuel Type

# Group data by primary_fuel and calculate mean, median, std, count of capacity_mw
group_stats = df.groupby('primary_fuel')['capacity_mw'].agg(['mean', 'median', 'std', 'count']).sort_values(by='mean', ascending=False)
print(group_stats)
```

```python
# Hypothesis Testing: Compare Mean Power Output Between Two Fuel Types

from scipy.stats import ttest_ind

# Extract capacity data for Coal and Solar power plants
coal_capacity = df[df['primary_fuel'] == 'Coal']['capacity_mw'].dropna()
solar_capacity = df[df['primary_fuel'] == 'Solar']['capacity_mw'].dropna()

# Perform two-sample t-test (Welch's t-test: unequal variances)
t_stat, p_value = ttest_ind(coal_capacity, solar_capacity, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4e}")

# Interpretation
if p_value < 0.05:
    print("There is a statistically significant difference in mean capacity between Coal and Solar power plants.")
else:
    print("No statistically significant difference in mean capacity between Coal and Solar power plants.")
```

```python
# Filter out rows without commissioning year (year == 0)
df_time = df[df['commissioning_year'] > 0].copy()

print(f"Data points with valid commissioning year: {len(df_time)}")
```

```python
#Time Series Analysis - Numerical Trends Over Years Using NumPy

import numpy as np

# Extract unique commissioning years and sort them
years = np.sort(df_time['commissioning_year'].unique())

# Count number of power plants commissioned each year using NumPy
counts_per_year = np.array([np.sum(df_time['commissioning_year'] == year) for year in years])

# Calculate rolling mean with window size 5 to smooth the trend
window = 5
rolling_mean = np.convolve(counts_per_year, np.ones(window)/window, mode='valid')

print("Years (first 10):", years[:10])
print("Counts per year (first 10):", counts_per_year[:10])
print("Rolling mean (5 years) (first 10):", rolling_mean[:10])
```

```python
# Exercise: Fuel Type Trends Over Years Using NumPy

# Get unique fuel types
fuel_types = df_time['primary_fuel'].unique()
fuel_trends = {}

# Calculate yearly counts of power plants for each fuel type
for fuel in fuel_types:
    counts = np.array([np.sum((df_time['commissioning_year'] == year) & (df_time['primary_fuel'] == fuel)) for year in years])
    fuel_trends[fuel] = counts

# Example: print trends for the first 3 fuel types
for fuel in list(fuel_types)[:3]:
    print(f"\nFuel type: {fuel}")
    print(f"Counts per year (first 10 years): {fuel_trends[fuel][:10]}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# We assume df_time contains cleaned data with valid 'commissioning_year' > 0
# The DataFrame includes 'longitude', 'latitude', and 'primary_fuel' columns

# Take a random sample of 5000 power plants for faster visualization
sample_df = df_time.sample(5000, random_state=42)

# Encode primary fuel types as numeric codes for coloring
sample_df['fuel_code'] = sample_df['primary_fuel'].astype('category').cat.codes
fuel_labels = sample_df['primary_fuel'].astype('category').cat.categories

# Create scatter plot
plt.figure(figsize=(14, 10))

scatter = plt.scatter(sample_df['longitude'], sample_df['latitude'],
                      c=sample_df['fuel_code'], cmap='tab20', alpha=0.6, s=15)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Power Plants by Fuel Type')

# Add colorbar with fuel type labels
cbar = plt.colorbar(scatter, ticks=range(len(fuel_labels)))
cbar.ax.set_yticklabels(fuel_labels)
cbar.set_label('Fuel Type')

plt.grid(True)
plt.show()

```

```python
# Extract unique sorted years
years = np.sort(df_time['commissioning_year'].unique())

# Calculate total number of power plants commissioned each year
counts_per_year = np.array([np.sum(df_time['commissioning_year'] == year) for year in years])

# Rolling mean to smooth the trend (window = 5 years)
window = 5
rolling_mean = np.convolve(counts_per_year, np.ones(window)/window, mode='valid')

# Plot total number of power plants commissioned per year
plt.figure(figsize=(14, 6))
plt.plot(years, counts_per_year, label='Annual Count', color='lightblue', alpha=0.7)
plt.plot(years[window-1:], rolling_mean, label='5-Year Rolling Mean', color='darkblue')
plt.xlabel('Commissioning Year')
plt.ylabel('Number of Power Plants')
plt.title('Number of Power Plants Commissioned Over Time')
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------------------------------------

# Fuel type trends over years

# Get unique fuel types
fuel_types = df_time['primary_fuel'].unique()

# Prepare color map for consistency
colors = plt.cm.tab20.colors

plt.figure(figsize=(14, 8))

# Plot trends for each fuel type
for i, fuel in enumerate(fuel_types):
    counts = np.array([np.sum((df_time['commissioning_year'] == year) & (df_time['primary_fuel'] == fuel)) for year in years])

    if counts.sum() > 0:  # Plot only if there are data points
        plt.plot(years, counts, label=fuel, color=colors[i % len(colors)])

plt.xlabel('Commissioning Year')
plt.ylabel('Number of Power Plants')
plt.title('Evolution of Power Plant Fuel Mix Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter dataset to relevant numerical attributes
# We will use latitude, longitude, and capacity for matrix operations
matrix_data = df[['latitude', 'longitude', 'capacity_mw']].dropna()

# Convert to NumPy matrix
X = matrix_data.to_numpy()

# Center the data by subtracting the mean
X_centered = X - X.mean(axis=0)

# Calculate covariance matrix
cov_matrix = np.cov(X_centered, rowvar=False)

print("Covariance matrix:\n", cov_matrix)

# Eigen decomposition of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors (columns correspond to principal directions):\n", eigenvectors)
```

Matrix & Eigenanalysis Findings:

✅ Most variability in the data comes from capacity differences between power plants.

✅ Geographic location (latitude & longitude) affects variability but much less.

✅ The first principal component (eigenvector) is strongly linked to capacity, showing it's the dominant factor.

✅ Useful for dimensionality reduction, identifying key trends, and simplifying complex spatial-capacity relationships.

```python
# Generate synthetic data using NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, size=100)  # sine wave with noise

# Moving average using NumPy
window_size = 5
moving_avg = np.convolve(y, np.ones(window_size)/window_size, mode='valid')

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Raw Data', alpha=0.6)
plt.plot(x[window_size-1:], moving_avg, color='red', label='Smoothed (Moving Avg)')
plt.title('Sine Wave with Noise and Smoothed Trend')
plt.legend()
plt.show()
```
