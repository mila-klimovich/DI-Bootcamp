```python
import numpy as np
import scipy.stats as stats

daily_steps = np.random.randint(3000, 10000, size=50)

mean_np = np.mean(daily_steps)
median_np = np.median(daily_steps)
variance_np = np.var(daily_steps)
std_dev_np = np.std(daily_steps)



mean_scipy = stats.tmean(daily_steps)
median_scipy = np.median(daily_steps)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(daily_steps)
std_dev_scipy = stats.tstd(daily_steps)

print("Mean:     NumPy =", mean_np, " | SciPy =", mean_scipy)
print("Median:   NumPy =", median_np)
print("Variance: NumPy =", variance_np, " | SciPy =", variance_scipy)
print("Std Dev:  NumPy =", std_dev_np, " | SciPy =", std_dev_scipy)

skewness = stats.skew(daily_steps)
kurtosis = stats.kurtosis(daily_steps)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate data
daily_steps = np.random.randint(4000, 10000, size=50)

# Compute statistics
mean = np.mean(daily_steps)
median = np.median(daily_steps)
skewness = stats.skew(daily_steps)
kurtosis = stats.kurtosis(daily_steps)

# Kernel Density Estimate for smooth curve
kde = stats.gaussian_kde(daily_steps)
x_vals = np.linspace(0, 15000, 500)
y_vals = kde(x_vals)

# Plot the KDE curve
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, color='blue')
plt.fill_between(x_vals, y_vals, color='blue', alpha=0.2)

# Add mean and median lines
plt.axvline(mean, color='red', linestyle='--', linewidth=1.5, label=f"Mean: {mean:.2f}")
plt.axvline(median, color='green', linestyle='--', linewidth=1.5, label=f"Median: {median:.2f}")

# Add text for skewness and kurtosis
plt.text(mean + 500, max(y_vals) * 0.8, f"Skewness: {skewness:.4f}\nKurtosis: {kurtosis:.4f}",
         fontsize=12, color='black')

# Title and labels
plt.title("Statistical Analysis of Daily Steps", fontsize=14)
plt.xlabel("Daily Steps")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate random data (simulate an experiment)
np.random.seed(42)  # Makes results repeatable

# Simulate blood pressure for 30 people in each group
control_group = np.random.normal(loc=125, scale=10, size=30)   # Placebo group
treatment_group = np.random.normal(loc=108, scale=10, size=30) # New drug group

# 2. Perform a T-test
t_stat, p_value = stats.ttest_ind(control_group, treatment_group)

# 3. Plot the histogram to visualize the results
plt.figure(figsize=(10, 6))

# Histograms: show distribution of blood pressure
plt.hist(control_group, bins=10, alpha=0.6, label='Control Group (Placebo)')
plt.hist(treatment_group, bins=10, alpha=0.6, label='Treatment Group (New Drug)')

# 4. Add dashed lines for the means of each group
plt.axvline(np.mean(control_group), color='blue', linestyle='dashed', linewidth=2, label='Mean Control')
plt.axvline(np.mean(treatment_group), color='orange', linestyle='dashed', linewidth=2, label='Mean Treatment')


# 5. Labels and title
plt.title('Blood Pressure Comparison Between Control and Treatment Groups')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.legend()

# 6. Show T-statistic and P-value on the plot
plt.text(90, 6.5, f'T-statistic: {t_stat:.3f}', fontsize=10)
plt.text(90, 6.0, f'P-value: {p_value:.2e}', fontsize=10)


```

```python
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(5,20)  # [5, 6, ..., 19]
y = np.exp(x/3.0)  # y = e^(x/3)
#Create Interpolation Function
f = interpolate.interp1d(x,y)

x1 = np.arange(6,12)        # New x values: 6, 7, 8, 9, 10, 11
y1 = f(x1)                # Estimate y-values for those x's
plt.plot(x,y, 'o' ,x1,y1, '--')
```
