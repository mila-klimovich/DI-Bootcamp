Exercise: Weather Data Analysis
Scenario:

Analyze temperature data across four cities over a week.

Data Set:

The temperature data for 4 cities (City 1 to City 4) over 7 days (Day 1 to Day 7)

Exercise Tasks:

1. Average Temperature Calculation: Compute the average temperature for each city over the week.

2. Maximum and Minimum Temperature Analysis: Find the highest and lowest temperatures recorded in each city.

3. Day Comparison: Compare the temperature across the cities on the first and last day of the week.

```python
import numpy as np

# Temperature data matrix (4 cities, 7 days)
temperature_data = np.array([
    [27, 20, 15, 18, 26, 18, 22],
    [24, 18, 20, 17, 19, 22, 21],
    [23, 23, 27, 25, 16, 21, 22],
    [22, 29, 23, 16, 20, 24, 28]
])

# Tasks: Calculate average, max, min temperatures and make day comparisons

avg_temp_per_city = np.mean(temperature_data, axis=1)
max_temp_per_city = np.max(temperature_data, axis=1)
min_temp_per_city = np.min(temperature_data, axis=1)

print("Average Temperatures per City:\n", avg_temp_per_city)
print("Maximum Temperatures per City:\n", max_temp_per_city)
print("Minimum Temperatures per City:\n", min_temp_per_city)

#compare Day 1 VS Day 7
day1 = temperature_data[:, 0]
day7 = temperature_data[:, 6]
print("Day 1 Temp:", day1)
print("Day 7 Temp:", day7)
print("Change from Day 1 to Day 7", day7 - day1)


max_day1 = np.max(day1)
max_day7 = np.max(day7)
min_day1 = np.min(day1)
min_day7 = np.min(day7)
print("Max temperature on Day 1:", max_day1)
print("Max temperature on Day 7:", max_day7)
print("Min temperature on Day 1:", min_day1)
print("Min temperature on Day 7:", min_day7)
```

Hypothesis Testing
What is Hypothesis Testing?

Hypothesis testing is a statistical method used to determine the likelihood of a hypothesis being true based on sample data. It involves making an initial assumption (the null hypothesis) and testing this assumption to infer about a population.

image.png

image.png

```python
import numpy as np
from scipy import stats

# Sample engagement times (in minutes)
sample_data = np.array([17, 18, 21, 20, 16, 19, 22, 20, 18, 21,
                        17, 20, 19, 18, 21, 20, 19, 22, 18, 19,
                        20, 21, 20, 19, 21, 20, 22, 19, 18, 21])

# Step 1: Test against H₀: μ = 15
t_statistic, p_value = stats.ttest_1samp(sample_data, 15)

# Step 2: Define significance level
alpha = 0.05

# Step 3: Decision
print(f"T-statistic: {t_statistic:.2f}, P-value: {p_value:.4f}")
if p_value < alpha:
    print("Reject the null hypothesis: Engagement time hasn't increased.")
else:
    print("Fail to reject the null hypothesis: Not enough evidence of increase.")
```

How spread out your data is (Variance)

Whether the data is symmetrical or skewed (Skewness)

Whether it has heavy or light tails (Kurtosis)

Skewness = 0 → perfect symmetry (normal distribution)

Positive skew → tail on the right (more high values)

Negative skew → tail on the left (more low values)

```python
from scipy.stats import skew

data = np.array([2, 4, 6, 8, 50, 1000])  # Extreme high value
skewness = skew(data)
print("Skewness:", skewness)
```

Kurtosis measures the "tailedness" of the distribution (how much data is in the tails vs. the center).

Kurtosis ≈ 3 → normal distribution (mesokurtic)

Kurtosis > 3 → heavy tails (leptokurtic) → more outliers

Kurtosis < 3 → light tails (platykurtic) → fewer outliers

```python
from scipy.stats import kurtosis

data = np.array([3, 5, 7, 9, 100, 120, 160])
kurt = kurtosis(data)  # Excess kurtosis
print("Excess Kurtosis:", kurt)
```

Time Series Use Case with Pandas + NumPy


While NumPy handles low-level date math efficiently, Pandas provides higher-level time series tools.

```python
import numpy as np
import pandas as pd

# Step 1: Generate a time series with 30 random values and daily date index
ts = pd.Series(np.random.randn(30), index=pd.date_range('2023-01-01', periods=30))
#print (ts)


# Step 2: Apply a 5-day rolling window and calculate the mean for each window
rolling_avg = ts.rolling(window=5).mean()

# Step 3: Print the result
print("\n5-Day Rolling Average:\n", rolling_avg)
```

```python
import pandas as pd

# Create a time series with 100 random values and date index
ts_data = pd.Series(np.random.randn(100), index=pd.date_range('2023-01-01', periods=100))

# Rolling mean over a 5-day window
rolling_mean = ts_data.rolling(window=5).mean()
print("Rolling Mean:\n", rolling_mean)

# Window sum
window_sum = ts_data.rolling(window=5).sum()
print("Window Sum:\n", window_sum)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


dates = np.arange('2022-01-01', '2023-01-01', dtype='datetime64[D]')
temperature = np.random.uniform(10, 35, size=len(dates))
humidity = np.random.uniform(30, 90, size=len(dates))

df = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

df.head()

```

```python
# Set Date as index
df.set_index('Date', inplace=True)

# Summary statistics
print("Temperature - Mean:", df['Temperature'].mean())
print("Temperature - Median:", df['Temperature'].median())
print("Temperature - Std Dev:", df['Temperature'].std())
print("Humidity - Mean:", df['Humidity'].mean())
print("Humidity - Median:", df['Humidity'].median())
print("Humidity - Std Dev:", df['Humidity'].std())
```

```python
df[['Temperature', 'Humidity']].plot(
    figsize=(12, 6),
    title='Daily Temperature and Humidity - 2022'
)
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
df['Temp_7day_avg'] = df['Temperature'].rolling(window=7).mean()
df['Hum_7days_avg'] = df['Humidity'].rolling(window=7).mean()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temperature'], alpha=0.4, label='Temperature (Daily)')
plt.plot(df.index, df['Temp_7day_avg'], color='red', label='Temp (7-day Avg)')
plt.plot(df.index, df['Humidity'], alpha=0.4, label='Humidity (Daily)')
plt.plot(df.index, df['Hum_7days_avg'], color='blue', label='Humidity (7-day Avg)')
plt.title("Temperature and Humidity with 7-Day Rolling Averages")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt

# ---------------------------------------------
# Step 1: Create a large DataFrame (100,000 rows × 4 columns)
# ---------------------------------------------
data = pd.DataFrame(np.random.rand(100000, 4), columns=['A', 'B', 'C', 'D'])

# ---------------------------------------------
# Step 2: Time Pandas operation
# ---------------------------------------------
start_time = time.time()
result_pandas = data[(data['A'] > 0.5) & (data['B'] < 0.5)].mean()
pandas_time = time.time() - start_time

print("Pandas Conditional Mean:\n", result_pandas)
print("Pandas Operation Time: {:.5f} seconds".format(pandas_time))

# ---------------------------------------------
# Step 3: Time NumPy equivalent
# ---------------------------------------------
numpy_data = data.to_numpy()
start_time = time.time()
result_numpy = np.mean(numpy_data[(numpy_data[:, 0] > 0.5) & (numpy_data[:, 1] < 0.5)], axis=0)
numpy_time = time.time() - start_time

print("NumPy Conditional Mean:\n", result_numpy)
print("NumPy Operation Time: {:.5f} seconds".format(numpy_time))

# ---------------------------------------------
# Step 4: Compare performance
# ---------------------------------------------
speed_ratio = pandas_time / numpy_time
print("NumPy is {:.2f}x faster than Pandas".format(speed_ratio))
```

```python
labels = ['A', 'B', 'C', 'D']
x = np.arange(len(labels))

plt.figure(figsize=(8, 6))
plt.bar(x, result_pandas.values, width=0.3, label='Pandas', color='skyblue')

plt.xticks(x, labels)
plt.ylabel("Mean Value")
plt.title("Conditional Mean Comparison: Pandas vs NumPy")
plt.legend()
plt.grid(axis='y')
plt.show()
```
