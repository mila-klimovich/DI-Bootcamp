```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

image_path = 'Flower Color Images Dataset/flowers/flowers/19_010.png'
original_image = Image.open(image_path)


horizontally_flipped = ImageOps.mirror(original_image)
vertically_flipped = ImageOps.flip(original_image)

width, height = original_image.size
zoomed_image = original_image.resize((int(width * 1.2), int(height * 1.2)))

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(original_image)
plt.title("Original")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(horizontally_flipped)
plt.title("Flipped Horizontally")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(vertically_flipped)
plt.title("Flipped Vertically")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(zoomed_image)
plt.title("Zoomed 1.2x")
plt.axis('off')

plt.tight_layout()
plt.show()
```
