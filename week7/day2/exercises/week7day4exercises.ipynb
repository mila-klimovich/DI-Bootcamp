Categorize each column of these datasets as either quantitative or qualitative.

Quantitative data are numbers you can measure or calculate with. Qualitative data describe categories or groups. For that reasoning:


For sleep_data dataset
Quantitative (numerical) columns:
index — numeric ID
Year — numeric
Avg hrs per day sleeping — numeric measurement
Standard Error — numeric statistical value

Qualitative (categorical) columns:
Period — category of time
Type of Days — day type
Age Group — age categories
Activity — type of activity
Sex — gender category


For mental_data dataset
Quantitative (numerical) columns:
index — numeric ID
Anxiety disorders — numeric (% values)
Drug use disorders — numeric (% values)
Depression — numeric (% values)
Alcohol use disorders — numeric (% values)

Qualitative (categorical) columns:
Entity — category
Code — category
Year — categorical here (object) can be numeric if converted

```python
import pandas as pd
import numpy as np

iris_data = pd.read_csv('Iris Dataset/Iris.csv')
print(iris_data.head(6))
```

```python
iris_data.head()
```

```python
iris_data.describe()
```

```python
iris_data.info()
```

Quantitative columns:
Id — A numeric identifier
SepalLengthCm — Numeric measurement (length in cm).
SepalWidthCm — Numeric measurement (width in cm).
PetalLengthCm — Numeric measurement (length in cm).
PetalWidthCm — Numeric measurement (width in cm).
These columns represent measurable physical dimensions that can be used in calculations and statistical analysis.

Qualitative column:
Species — Text label indicating the flower species.
It describes a category or group and cannot be used in mathematical operations.

```python
mean_pw = iris_data['PetalWidthCm'].mean()
print('Mean Petal Width:', mean_pw)

median_pw = iris_data['PetalWidthCm'].median()
print('Median Petal Width:', median_pw)

mode_pw = iris_data['PetalWidthCm'].mode()[0]
print('Mode Petal Width:', mode_pw)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(iris_data['PetalWidthCm'], bins=20, color='lightgreen')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

Findings:

- Mean: 1.20 cm  
- Median: 1.30 cm  
- Mode: 0.2 cm  

The histogram shows a bimodal distribution, suggesting two or more dominant petal width groups — likely corresponding to different species in the dataset.
