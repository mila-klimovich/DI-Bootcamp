Exercise 3.
Categorize each column of these datasets as either quantitative or qualitative.

Quantitative data are numbers you can measure or calculate with. Qualitative data describe categories or groups. For that reasoning:


For sleep_data dataset
Quantitative (numerical) columns:
index — numeric ID
Year — numeric
Avg hrs per day sleeping — numeric measurement
Standard Error — numeric statistical value

Qualitative (categorical) columns:
Period — category of time
Type of Days — day type
Age Group — age categories
Activity — type of activity
Sex — gender category


For mental_data dataset
Quantitative (numerical) columns:
index — numeric ID
Anxiety disorders — numeric (% values)
Drug use disorders — numeric (% values)
Depression — numeric (% values)
Alcohol use disorders — numeric (% values)

Qualitative (categorical) columns:
Entity — category
Code — category
Year — categorical here (object) can be numeric if converted
Schizophrenia (%), Bipolar disorder (%), Eating disorders (%) — stored as objects, so categorical unless converted to numeric
