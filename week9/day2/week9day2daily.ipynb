```python
print(df.isnull().sum())
df = df.dropna()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df['Social support'], df['Score'], color='teal')
plt.title('Relationship between Social Support and Happiness Score')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```

```python
# For demonstration, we'll assign dummy regions
import numpy as np

# Example dummy regions based on index (replace with actual logic if available)
df['Region'] = np.where(df.index < 50, 'Europe',
                        np.where(df.index < 100, 'Asia', 'Other'))

# Group by Region
region_group = df.groupby('Region').agg({
    'GDP per capita': 'mean',
    'Healthy life expectancy': 'mean'
}).reset_index()

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for GDP per capita
ax1.bar(region_group['Region'], region_group['GDP per capita'], color='skyblue', label='GDP per Capita')
ax1.set_ylabel('GDP per Capita', color='blue')
ax1.set_xlabel('Region')
ax1.tick_params(axis='y', labelcolor='blue')

# Line plot for Healthy life expectancy
ax2 = ax1.twinx()
ax2.plot(region_group['Region'], region_group['Healthy life expectancy'], color='darkgreen', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy Life Expectancy', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('GDP per Capita and Healthy Life Expectancy by Region')
plt.grid(False)
plt.tight_layout()
plt.show()
```
