Why is data visualization important in data analysis?
Data visualization helps us understand complex data more easily by turning numbers into charts and graphs. It makes patterns, trends, and outliers easier to spot, which helps in making better decisions and communicating findings clearly to others.

What is the purpose of a line graph in data visualization?
A line graph is used to show how data changes over time or continuous intervals. It helps track trends, increases, or decreases clearly, making it easy to understand how something evolves.

```python
import matplotlib.pyplot as plt

temperatures = [72, 74, 76, 80, 82, 78, 75]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.plot(days, temperatures, marker='o', linestyle='-', color='b')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°F)')
plt.title('Temperature Variation Over a Week')

plt.show()
```

```python
import matplotlib.pyplot as plt

sales = [5000, 5500, 6200, 7000, 7500, 8000, 8200, 7800, 7600, 8100, 8500, 9000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.bar(months, sales, color='skyblue')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales for Retail Store')

plt.show()
```

```python
print(df.isnull().sum())
```

```python
total_quantity = df['quantity'].sum()
print(f"Total quantity sold: {total_quantity}")
```

```python
category_revenue = df.groupby('category')['revenue'].sum()

max_category = category_revenue.idxmax()
max_revenue = category_revenue.max()

print(f"Category with highest revenue: {max_category} (${max_revenue:.2f})")
```

```python
average_revenue_per_sale = df['revenue'].mean()
print(f"Average revenue per sale: ${average_revenue_per_sale:.2f}")
```

```python
import matplotlib.pyplot as plt

df['date'] = pd.to_datetime(df['date'], errors='coerce')

df['quarter'] = df['date'].dt.to_period('Q')

quarter_revenue = df.groupby('quarter')['revenue'].sum()

quarter_revenue.plot(kind='bar', color='orange')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.title('Total Revenue by Quarter')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1) Line plot for y = x^2
x = np.arange(-10, 11)  # x values from -10 to 10
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', color='blue')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# 2) Bar chart for sales data of products A, B, C, D
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure(figsize=(6, 4))
plt.bar(products, sales, color='purple')
plt.title('Weekly Sales of Products')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# 3) Pie chart for favorite fruits data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [40, 30, 20, 10]
colors = ['red', 'yellow', 'pink', 'brown']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Favorite Fruits Distribution')
plt.legend(labels, loc='best')
plt.show()
```
