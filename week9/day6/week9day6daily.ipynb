```python
import pandas as pd

# Load the Excel file (usually there's only one sheet)
df = pd.read_excel('US Superstore data.xls')

# Show the first few rows
df.head()
```

```python
# Shape of the dataset
print("Rows and Columns:", df.shape)

# Check column names
print("\nColumn Names:\n", df.columns)

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Basic statistics
print("\nSummary Stats:\n", df.describe(include='all'))
```

```python
print(f"Duplicates: {df.duplicated().sum()}")
df = df.drop_duplicates()
```

```python
# Ð¡Ð¾Ð·Ð´Ð°ÐµÐ¼ Ð½Ð¾Ð²Ñ‹Ðµ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¸
df['Shipping Delay'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

# ÐžÑ‡Ð¸Ñ‰Ð°ÐµÐ¼ ÐºÐ°Ñ‚ÐµÐ³Ð¾Ñ€Ð¸Ð°Ð»ÑŒÐ½Ñ‹Ðµ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¸ Ð¾Ñ‚ Ð¿Ñ€Ð¾Ð±ÐµÐ»Ð¾Ð²
cat_cols = ['Ship Mode', 'Segment', 'Country', 'Region', 'Category', 'Sub-Category']
for col in cat_cols:
    df[col] = df[col].str.strip()
```

```python
import plotly.express as px

# Aggregate sales by year
sales_by_year = df.groupby('Order Year')['Sales'].sum().reset_index()

# Create interactive line chart
fig = px.line(sales_by_year, x='Order Year', y='Sales', title='Sales Trends Over the Years')
fig.show()
```

```python
import us

def get_state_abbr(name):
    try:
        return us.states.lookup(name).abbr
    except:
        return None

df['State_Abbr'] = df['State'].apply(get_state_abbr)
```

```python
print(df[df['State_Abbr'].isnull()][['State']].drop_duplicates())
# Fix missing abbreviation for 'District of Columbia'
df.loc[df['State'] == 'District of Columbia', 'State_Abbr'] = 'DC'
print(df[df['State_Abbr'].isnull()][['State']].drop_duplicates())
```

```python
import plotly.express as px

sales_by_state = df.groupby('State_Abbr')['Sales'].sum().reset_index()

fig = px.choropleth(
    sales_by_state,
    locations='State_Abbr',
    locationmode='USA-states',
    color='Sales',
    scope='usa',
    color_continuous_scale='Blues',
    title='Sales Distribution by US State'
)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate sales by product
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=top_products, x='Sales', y='Product Name', palette='viridis')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.6)
plt.title('Profit vs Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

Comparative Analysis: Matplotlib vs. Seaborn

Ease of Use:

Matplotlib:

Very flexible and powerful â€” allows precise control over every element of the plot.

However, it can require more lines of code and detailed parameter tuning to get polished visuals.

Good for simple plots and when you want full customization.

Seaborn:

Built on top of Matplotlib but designed for statistical graphics.

Provides high-level functions with smart defaults, which means less code to produce aesthetically pleasing and informative plots.

Easily handles complex visualizations like grouped bar charts, scatter plots with regression lines, heatmaps, etc.

Effectiveness for Your Tasks:

Matplotlib Sales Trend Line Chart:

Great for showing changes over time with full control of formatting axes, labels, and layout.

Interactive widgets can be added but require extra setup (e.g., %matplotlib widget or ipywidgets).

Seaborn Bar and Scatter Plots:

Quickly generated polished charts ideal for exploratory data analysis.

Easily handles aggregation and categorization, which is useful for â€œtop 10 productsâ€ or relationship plots.

Better default color palettes and styles make charts visually appealing without much tweaking.

Summary:

For quick, beautiful statistical plots with less boilerplate, Seaborn is the winner.

For fine-tuned, highly customized plots or when interactive control is needed, Matplotlib offers more flexibility.

Often the best workflow is to start with Seaborn for EDA, then switch to Matplotlib when you need to polish the details.
