```python
mall_data = pd.read_csv('Mall_Customers.csv')
mall_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(mall_data['Annual_Income_(k$)'], bins=20, color='skyblue')
plt.title('Histogram of Annual Income')
plt.grid(True)
# plt.show()
```

```python
sns.kdeplot(mall_data['Annual_Income_(k$)'], shade=True, color='red')
plt.title('KDE of Annual Income')
# plt.show()
plt.grid(True)
```

```python
sns.kdeplot(mall_data['Age'], bw_adjust=0.5, color='green', shade=True)
plt.title('Customized KDE of Age')
plt.show()
plt.grid(True)
```
