```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])
df[['salary', 'salary_normalized']].head()
```

```python
df_encoded = pd.get_dummies(df.drop(columns=['Unnamed: 0', 'salary_normalized']), drop_first=True)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_encoded)

df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca.head()
```

```python
grouped_salary = df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()
grouped_salary.rename(columns={'mean': 'avg_salary', 'median': 'median_salary'}, inplace=True)
print(grouped_salary)
```
