```python
import pandas as pd

df1 = pd.read_csv("Bigtech - 12-07-2020 till 19-09-2020.csv")
df2 = pd.read_csv("Bigtech - 20-09-2020 till 13-10-2020.csv")

print("DF1 preview:")
display(df1.head())

print("DF2 preview:")
display(df2.head())
```

```python
print("DF1 columns:")
print(df1.columns)

print("\nDF2 columns:")
print(df2.columns)

print(f"\nShape DF1: {df1.shape}")
print(f"Shape DF2: {df2.shape}")
```

```python
# Combine both datasets into one
df = pd.concat([df1, df2], ignore_index=True)

# Display the shape and first few rows
print("Combined dataset shape:", df.shape)
df.head()
```

```python
# Display the first few rows of the merged dataset
df.head()
```

```python
# Check basic info about data types and missing values
df.info()
```

```python
#Missing Values Analysis

# Count missing values per column
missing_values = df.isnull().sum()

# Display only columns with missing values
missing_values = missing_values[missing_values > 0]

# Sort by number of missing values
missing_values.sort_values(ascending=False)
```

```python
# Summary statistics for numerical columns
df.describe()
```

The combined dataset contains 1,133,004 tweets with 15 columns, including tweet metadata, user information, and sentiment polarity scores.

Key features include created_at (tweet timestamp), text (tweet content), followers and friends (user metrics), retweet_count (engagement), group_name and partition_1 (company names), and polarity (sentiment score ranging from -1 to 1).

Missing values are minimal except for the location field, which has about 28% missing data.

Numerical features like followers and retweet_count have wide distributions with large maximum values, indicating the presence of outliers.

The sentiment polarity ranges from approximately -1 (negative) to +1 (positive) with a median near zero, suggesting a balanced distribution of sentiments.

Overall, the dataset is suitable for further analysis with minor preprocessing required to handle missing values and outliers.

```python
# Sentiment Polarity Distribution - full view with mean, median, and skewness

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Drop NaNs in polarity
polarity_values = df['polarity'].dropna()

# Calculate mean, median, and skewness
mean_polarity = polarity_values.mean()
median_polarity = polarity_values.median()
skewness_polarity = skew(polarity_values)

# Set visual style
sns.set(style="whitegrid")

# Plot histogram
plt.figure(figsize=(10,6))
sns.histplot(polarity_values, bins=50, kde=True, color='skyblue')

# Add lines for mean and median
plt.axvline(x=0, color='red', linestyle='--', label='Neutral Sentiment')
plt.axvline(x=mean_polarity, color='green', linestyle='--', label=f'Mean = {mean_polarity:.2f}')
plt.axvline(x=median_polarity, color='purple', linestyle='--', label=f'Median = {median_polarity:.2f}')

plt.title("Distribution of Tweet Polarity")
plt.xlabel("Polarity Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Print skewness
print(f"Skewness of Polarity: {skewness_polarity:.2f}")
```

The polarity distribution is right-skewed, indicating that most tweets have neutral to slightly positive sentiment.

The mean polarity is slightly above zero, confirming a generally positive trend.

The median is close to zero, suggesting that neutral sentiment dominates.

A small number of highly negative tweets pull the left tail down.

This suggests that public discussions about tech companies tend to lean positive, though strong negative opinions also exist, especially as outliers.

```python
# Sentiment by Tech Company - Compare how public sentiment (polarity) varies across different tech companies mentioned in the tweets.
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='partition_1', y='polarity', palette='Set2')
plt.title('Sentiment Polarity by Tech Company')
plt.xlabel('Tech Company')
plt.ylabel('Polarity')
plt.xticks(rotation=45)
plt.show()
```

The boxplot illustrates that public sentiment on Twitter is generally positive for all the tech companies analyzed. Twitch and Netflix show the highest median polarity, suggesting favorable public opinion. In contrast, Google and Nvidia have slightly lower sentiment medians. All companies exhibit a wide sentiment range, with some negative outliers, indicating the presence of critical tweets or negative feedback.

```python
# Analyze the relationship between user influence (follower count) and sentiment polarity - to Explore whether users with higher follower counts express stronger or more polarized opinions

# Create follower count groups for better comparison
df['follower_group'] = pd.cut(
    df['followers'],
    bins=[-1, 100, 1000, 10000, 100000, 1000000, df['followers'].max()],
    labels=['0-100', '101-1K', '1K-10K', '10K-100K', '100K-1M', '1M+']
)

# Visualize polarity distribution by follower group
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='follower_group', y='polarity', palette='Spectral')
plt.title('Polarity by Follower Group')
plt.xlabel('Follower Count Group')
plt.ylabel('Sentiment Polarity')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The boxplot indicates that overall sentiment polarity does not vary drastically between follower count groups. However, users with fewer followers (0–100) tend to have slightly more positive median sentiment compared to high-follower accounts (1M+), which show a slightly lower and more varied sentiment distribution. This may suggest that high-profile users are more neutral or cautious in tone, or face more polarized reactions.

```python
# Analyze Tweet Activity Over Time
# Convert 'created_at' column to datetime
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')

# Extract only the date (without time)
df['tweet_date'] = df['created_at'].dt.date

# Count number of tweets per day
daily_tweet_counts = df['tweet_date'].value_counts().sort_index()

# Plot tweet activity over time
plt.figure(figsize=(16, 6))
sns.lineplot(x=daily_tweet_counts.index, y=daily_tweet_counts.values)
plt.title('Tweet Volume Over Time', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The volume of tweets saw a significant spike in mid-July 2020, reaching over 20,000 tweets per day. After that, tweet activity stabilized but remained consistently high throughout the observation period.

There is also a notable drop in mid-September, possibly due to data collection issues or temporary decline in public discussions.

These fluctuations may correspond to tech-related announcements, product releases, or controversies, which can be explored further by aligning dates with real-world events.

```python
#Geographic Analysis of Tweet Locations
# Check most common locations in the dataset
top_locations = df['location'].value_counts().dropna().head(20)

# Plot top locations by frequency
plt.figure(figsize=(12,6))
sns.barplot(y=top_locations.index, x=top_locations.values, palette="viridis")
plt.title("Top 20 Most Frequent Tweet Locations")
plt.xlabel("Number of Tweets")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

The majority of tweets originate from the **United States**, followed by **India**, **United Kingdom**, and several major cities like **Los Angeles**, **New York**, and **London**.

This suggests that public sentiment on tech companies is primarily shaped by users from these regions. The presence of generic or inconsistent entries like "Worldwide" and missed values may require further cleaning for advanced geographic analysis.
