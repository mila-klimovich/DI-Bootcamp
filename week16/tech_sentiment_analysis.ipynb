```python
import pandas as pd

df1 = pd.read_csv("Bigtech - 12-07-2020 till 19-09-2020.csv")
df2 = pd.read_csv("Bigtech - 20-09-2020 till 13-10-2020.csv")

print("DF1 preview:")
display(df1.head())

print("DF2 preview:")
display(df2.head())
```

```python
print("DF1 columns:")
print(df1.columns)

print("\nDF2 columns:")
print(df2.columns)

print(f"\nShape DF1: {df1.shape}")
print(f"Shape DF2: {df2.shape}")
```

```python
# Combine both datasets into one
df = pd.concat([df1, df2], ignore_index=True)

# Display the shape and first few rows
print("Combined dataset shape:", df.shape)
df.head()
```

```python
# Display the first few rows of the merged dataset
df.head()
```

```python
# Check basic info about data types and missing values
df.info()
```

```python
#Missing Values Analysis

# Count missing values per column
missing_values = df.isnull().sum()

# Display only columns with missing values
missing_values = missing_values[missing_values > 0]

# Sort by number of missing values
missing_values.sort_values(ascending=False)
```

```python
# Summary statistics for numerical columns
df.describe()
```

The combined dataset contains 1,133,004 tweets with 15 columns, including tweet metadata, user information, and sentiment polarity scores.

Key features include created_at (tweet timestamp), text (tweet content), followers and friends (user metrics), retweet_count (engagement), group_name and partition_1 (company names), and polarity (sentiment score ranging from -1 to 1).

Missing values are minimal except for the location field, which has about 28% missing data.

Numerical features like followers and retweet_count have wide distributions with large maximum values, indicating the presence of outliers.

The sentiment polarity ranges from approximately -1 (negative) to +1 (positive) with a median near zero, suggesting a balanced distribution of sentiments.

Overall, the dataset is suitable for further analysis with minor preprocessing required to handle missing values and outliers.

```python
# Sentiment Polarity Distribution - full view with mean, median, and skewness

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Drop NaNs in polarity
polarity_values = df['polarity'].dropna()

# Calculate mean, median, and skewness
mean_polarity = polarity_values.mean()
median_polarity = polarity_values.median()
skewness_polarity = skew(polarity_values)

# Set visual style
sns.set(style="whitegrid")

# Plot histogram
plt.figure(figsize=(10,6))
sns.histplot(polarity_values, bins=50, kde=True, color='skyblue')

# Add lines for mean and median
plt.axvline(x=0, color='red', linestyle='--', label='Neutral Sentiment')
plt.axvline(x=mean_polarity, color='green', linestyle='--', label=f'Mean = {mean_polarity:.2f}')
plt.axvline(x=median_polarity, color='purple', linestyle='--', label=f'Median = {median_polarity:.2f}')

plt.title("Distribution of Tweet Polarity")
plt.xlabel("Polarity Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Print skewness
print(f"Skewness of Polarity: {skewness_polarity:.2f}")
```

The polarity distribution is right-skewed, indicating that most tweets have neutral to slightly positive sentiment.

The mean polarity is slightly above zero, confirming a generally positive trend.

The median is close to zero, suggesting that neutral sentiment dominates.

A small number of highly negative tweets pull the left tail down.

This suggests that public discussions about tech companies tend to lean positive, though strong negative opinions also exist, especially as outliers.

```python
# Sentiment by Tech Company - Compare how public sentiment (polarity) varies across different tech companies mentioned in the tweets.
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='partition_1', y='polarity', palette='Set2')
plt.title('Sentiment Polarity by Tech Company')
plt.xlabel('Tech Company')
plt.ylabel('Polarity')
plt.xticks(rotation=45)
plt.show()
```

The boxplot illustrates that public sentiment on Twitter is generally positive for all the tech companies analyzed. Twitch and Netflix show the highest median polarity, suggesting favorable public opinion. In contrast, Google and Nvidia have slightly lower sentiment medians. All companies exhibit a wide sentiment range, with some negative outliers, indicating the presence of critical tweets or negative feedback.

```python
# Analyze the relationship between user influence (follower count) and sentiment polarity - to Explore whether users with higher follower counts express stronger or more polarized opinions

# Create follower count groups for better comparison
df['follower_group'] = pd.cut(
    df['followers'],
    bins=[-1, 100, 1000, 10000, 100000, 1000000, df['followers'].max()],
    labels=['0-100', '101-1K', '1K-10K', '10K-100K', '100K-1M', '1M+']
)

# Visualize polarity distribution by follower group
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='follower_group', y='polarity', palette='Spectral')
plt.title('Polarity by Follower Group')
plt.xlabel('Follower Count Group')
plt.ylabel('Sentiment Polarity')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The boxplot indicates that overall sentiment polarity does not vary drastically between follower count groups. However, users with fewer followers (0–100) tend to have slightly more positive median sentiment compared to high-follower accounts (1M+), which show a slightly lower and more varied sentiment distribution. This may suggest that high-profile users are more neutral or cautious in tone, or face more polarized reactions.

```python
# Analyze Tweet Activity Over Time
# Convert 'created_at' column to datetime
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')

# Extract only the date (without time)
df['tweet_date'] = df['created_at'].dt.date

# Count number of tweets per day
daily_tweet_counts = df['tweet_date'].value_counts().sort_index()

# Plot tweet activity over time
plt.figure(figsize=(16, 6))
sns.lineplot(x=daily_tweet_counts.index, y=daily_tweet_counts.values)
plt.title('Tweet Volume Over Time', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The volume of tweets saw a significant spike in mid-July 2020, reaching over 20,000 tweets per day. After that, tweet activity stabilized but remained consistently high throughout the observation period.

There is also a notable drop in mid-September, possibly due to data collection issues or temporary decline in public discussions.

These fluctuations may correspond to tech-related announcements, product releases, or controversies, which can be explored further by aligning dates with real-world events.

```python
# Data Preprocessing
# Check missing values again to confirm
missing_values = df.isnull().sum()
print("Missing values before handling:")
print(missing_values[missing_values > 0])
```

```python
# Drop rows with missing critical values
df = df.dropna(subset=['text', 'polarity', 'retweet_count', 'twitter_id'])

# Fill missing location and username with 'Unknown' if needed later
df['location'] = df['location'].fillna('Unknown')
df['username'] = df['username'].fillna('Unknown')
```

```python
# Check for duplicate tweets based on text and user
duplicates = df.duplicated(subset=['text', 'username'])
print(f"Number of duplicate tweets: {duplicates.sum()}")

# Drop duplicates
df = df.drop_duplicates(subset=['text', 'username'])
print(f"Shape after removing duplicates: {df.shape}")
```

```python
# Convert 'created_at' column to datetime
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')

# Check for any conversion issues
print(f"Number of NaT values in 'created_at': {df['created_at'].isna().sum()}")

# Preview updated column
df[['created_at']].head()
```

```python
# Boxplots for Outlier Detection
# Set visual style
sns.set(style="whitegrid")

# Define numeric columns to check for outliers
numeric_cols = ['followers', 'friends', 'retweet_count', 'polarity']

# Create boxplots for each variable
plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

Outlier Analysis

Followers: The majority of users have fewer than ~10,000 followers, but a small number of accounts exceed millions — strong outliers.

Friends: Similarly skewed, with most values concentrated near zero and a long tail of high values.

Retweet Count: Most tweets have 0–10 retweets, but a few viral posts have tens of thousands — extreme right skew.

Polarity: Distribution appears more symmetric, ranging from -1 (negative) to 1 (positive), with fewer outliers.

Conclusion:
Outliers are expected in social media data, especially in popularity-related metrics. Since no modeling is planned, we can retain the outliers to preserve real-world variability and avoid losing influential users or viral tweets.

```python
# Correlation and Trend Analysis

# Correlation Matrix of Numeric Variables
# Select numerical columns for correlation
numeric_cols = ['followers', 'friends', 'retweet_count', 'polarity']
correlation_matrix = df[numeric_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix of Numeric Variables")
plt.show()
```

Insights from the Correlation Matrix:

Followers vs Friends
A weak positive correlation of 0.14 suggests that users with more followers tend to have slightly more friends, but the relationship is not strong.

Retweet Count vs Other Variables
Retweet count has negligible correlation with followers, friends, and polarity, indicating that viral tweets are not necessarily influenced by user popularity or sentiment.

Polarity vs Other Variables
Polarity scores show almost no linear relationship with any of the numeric variables (all values around 0), implying that user influence or retweet activity does not significantly affect sentiment.

```python
# Time Trends in Sentiment - Average Polarity Over Time

# Convert 'created_at' to datetime if not already
df['created_at'] = pd.to_datetime(df['created_at'])

# Create daily average polarity
daily_polarity = df.groupby(df['created_at'].dt.date)['polarity'].mean()

# Plot
plt.figure(figsize=(12,6))
daily_polarity.plot()
plt.title("Average Sentiment Polarity Over Time")
plt.xlabel("Date")
plt.ylabel("Average Polarity")
plt.axhline(y=0, color='gray', linestyle='--', label='Neutral')
plt.legend()
plt.tight_layout()
plt.show()
```

 Sentiment Polarity Over Time

The average daily sentiment polarity consistently stays above 0, indicating an overall positive tone in public discussions about tech companies during the observed period. While there are some fluctuations, no major spikes or sentiment drops are evident, suggesting relatively stable sentiment throughout.


```python
# Combined Trend of Tweet Volume and Average Polarity Over Time

# Prepare data
daily_stats = df.groupby(df['created_at'].dt.date).agg({
    'text': 'count',
    'polarity': 'mean'
}).rename(columns={'text': 'tweet_count'})

# Plot
fig, ax1 = plt.subplots(figsize=(12,6))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('Tweet Count', color=color)
ax1.plot(daily_stats.index, daily_stats['tweet_count'], color=color, label='Tweet Count')
ax1.tick_params(axis='y', labelcolor=color)

# Create second y-axis
ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Average Polarity', color=color)
ax2.plot(daily_stats.index, daily_stats['polarity'], color=color, label='Avg Polarity')
ax2.axhline(y=0, color='gray', linestyle='--', linewidth=1)
ax2.tick_params(axis='y', labelcolor=color)

# Title and layout
plt.title('Tweet Volume vs. Sentiment Polarity Over Time')
fig.tight_layout()
plt.show()
```

Tweet Volume vs. Sentiment Polarity Over Time

This dual-axis plot shows both the daily number of tweets and the average sentiment polarity. A significant increase in tweet activity is observed in mid-July 2020, after which volume stabilizes around 10–15K tweets per day. Despite fluctuations in volume, the sentiment remains consistently positive throughout the time period, suggesting that public discourse around tech companies stayed optimistic even during spikes in activity.


Key Insights

- **Overall Sentiment:** The average tweet polarity across the dataset is slightly positive, with most tweets clustering around the neutral zone. Extreme sentiments (very positive or very negative) are less frequent.

- **Company Perception:** Among the analyzed tech companies, Twitch, Netflix, and Amazon had slightly more positive sentiment distributions, while Nvidia and Google showed relatively more neutral or mixed polarity.

- **User Influence & Sentiment:** Users with smaller follower bases (0–100) tend to express more positive sentiment than high-profile accounts (1M+), whose sentiment appears more neutral and varied.

- **Temporal Trends:** A clear spike in tweet volume occurred around mid-July 2020, possibly driven by an event or announcement. Despite fluctuations in volume, sentiment remained fairly consistent over time.

- **Geographic Trends:** Most tweets originated from the United States, India, and the United Kingdom. This highlights regional engagement in tech conversations, especially from English-speaking countries.

- **Correlation Analysis:** No strong linear correlations were found between numeric features such as follower count, retweets, and sentiment polarity. The variables are mostly independent.



Conclusion

This sentiment analysis of over 1.1 million tweets related to major tech companies reveals meaningful insights into public opinion across the tech industry.

The polarity distribution indicates that most tweets carry neutral or mildly positive sentiment, with only a small proportion expressing strong negativity or positivity. Brands like Twitch and Netflix received more consistently favorable sentiment, while others like Google and Nvidia exhibited a more neutral or mixed public perception.

User influence (measured by followers or retweets) does not significantly correlate with sentiment polarity, suggesting that both low- and high-profile users contribute similarly to the sentiment landscape.

Temporal analysis showed steady sentiment patterns over time, even during tweet volume spikes—highlighting the importance of context when interpreting public discussion trends.
