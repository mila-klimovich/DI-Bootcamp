```python
# Challenge: Analyze Global Weather Temperature Trends

# Step 1: Data Preparation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define parameters
num_cities = 10
num_months = 12
temp_low = -5
temp_high = 35

# Generate synthetic temperature data: 10 cities x 12 months
temps = np.random.uniform(temp_low, temp_high, size=(num_cities, num_months))

# Create city names and month names
cities = [f"City_{i+1}" for i in range(num_cities)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Create DataFrame
df_temps = pd.DataFrame(data=temps, index=cities, columns=months)

print("Temperature DataFrame (sample):")
print(df_temps.head())
```

```python
# Step 2: Data Analysis

# Calculate annual average temperature per city
df_temps['Annual_Avg'] = df_temps.mean(axis=1)

# Identify city with highest average temperature
city_highest_avg = df_temps['Annual_Avg'].idxmax()
highest_avg_temp = df_temps.loc[city_highest_avg, 'Annual_Avg']

# Identify city with lowest average temperature
city_lowest_avg = df_temps['Annual_Avg'].idxmin()
lowest_avg_temp = df_temps.loc[city_lowest_avg, 'Annual_Avg']

print(f"City with highest average temperature: {city_highest_avg} ({highest_avg_temp:.2f} °C)")
print(f"City with lowest average temperature: {city_lowest_avg} ({lowest_avg_temp:.2f} °C)")
```

```python
# Step 3: Data Visualization

plt.figure(figsize=(12, 7))

for city in cities:
    plt.plot(months, df_temps.loc[city, months], marker='o', label=city)

plt.title('Monthly Temperature Trends for Each City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

Summary Report
The dataset simulates average monthly temperatures for 10 cities over a year.

The city with the highest annual average temperature is City_5 with approximately 19.36 °C.

The city with the lowest annual average temperature is City_2 with approximately 10.74 °C.

The temperature trends vary across cities with some cities showing a wider range of temperature fluctuation, indicating stronger seasonality.

Visualization helps quickly identify which cities experience warmer or colder months and the relative stability or volatility in temperatures.
